```python
# CADL4: Topic Modeling with LDA

# Fix version conflicts by installing compatible versions
!pip install numpy==1.26.4 scipy==1.13.1 gensim==4.3.3 pyLDAvis==3.4.1

import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

import gensim
from gensim import corpora
import pyLDAvis.gensim_models
import pyLDAvis

# --- Example corpus
documents = [
    "Machine learning improves computer vision and artificial intelligence.",
    "Elections and politics dominate world news today.",
    "Football and cricket are the most popular sports worldwide.",
    "Deep learning and neural networks advance healthcare and medicine.",
    "Government policies affect economic growth and development.",
    "Basketball and Olympics bring nations together through sports."
]

# --- 1. Preprocessing ---
stop_words = stopwords.words('english')
texts = [[word for word in doc.lower().split() if word not in stop_words] for doc in documents]

# Create dictionary and corpus
dictionary = corpora.Dictionary(texts)
corpus = [dictionary.doc2bow(text) for text in texts]

# --- 2. Apply LDA (3 topics) ---
lda_model = gensim.models.LdaModel(
    corpus,
    num_topics=3,
    id2word=dictionary,
    passes=15,
    random_state=42
)

print("\nLDA Topics:")
for idx, topic in lda_model.print_topics(num_words=5):
    print(f"Topic {idx+1}: {topic}")

# --- 3. Visualization ---
pyLDAvis.enable_notebook()
vis = pyLDAvis.gensim_models.prepare(lda_model, corpus, dictionary)
vis
```
